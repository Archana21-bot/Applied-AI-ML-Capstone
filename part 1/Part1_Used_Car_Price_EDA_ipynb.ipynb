```python
null_count = df.isnull().sum()
null_percentage = (null_count / len(df)) * 100

null_df = pd.DataFrame({
    "Missing Values": null_count,
    "Percentage (%)": null_percentage.round(2)
})

print("Null Value Analysis")
display(null_df)

high_null = null_df[null_df["Percentage (%)"] > 20]

print("\nColumns with more than 20% missing values:")
if high_null.empty:
    print("None")
else:
    display(high_null)
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

rows_before = df.shape[0]
df = df.drop_duplicates()
rows_after = df.shape[0]

print(f"Rows removed: {rows_before - rows_after}")
print(f"New dataset shape: {df.shape}")

new_null_percentage = (df.isnull().sum() / len(df)) * 100

print("\nNull Percentage After Removing Duplicates")
display(new_null_percentage.round(2))
```

```python
df["milage"] = (
    df["milage"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(" mi.", "", regex=False)
)

df["milage"] = pd.to_numeric(df["milage"], errors="coerce")

print(df["milage"].dtype)
```

```python
for col in ["price", "milage"]:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"\n{col}")
    print(f"Q1: {Q1}")
    print(f"Q3: {Q3}")
    print(f"IQR: {IQR}")
    print(f"Lower Bound: {lower}")
    print(f"Upper Bound: {upper}")
    print(f"Number of Outliers: {len(outliers)}")
```

```python
numeric_cols = df.select_dtypes(include="number").columns

skewness = df[numeric_cols].skew().abs().sort_values(ascending=False)
top2 = skewness.index[:2]

comparison = pd.DataFrame({
    "Mean": df[top2].mean(),
    "Median": df[top2].median()
})

print("Mean vs Median Comparison")
display(comparison)

for col in top2:
    df[col].fillna(df[col].median(), inplace=True)

print("\nRemaining Missing Values")
print(df[top2].isnull().sum())
```

```python
pearson = df.select_dtypes(include="number").corr(method="pearson")
spearman = df.select_dtypes(include="number").corr(method="spearman")

print("Pearson Correlation")
display(pearson)

print("Spearman Correlation")
display(spearman)

pairs = []

cols = pearson.columns

for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        diff = abs(spearman.iloc[i, j] - pearson.iloc[i, j])
        pairs.append([
            cols[i],
            cols[j],
            pearson.iloc[i, j],
            spearman.iloc[i, j],
            diff
        ])

difference_table = pd.DataFrame(
    pairs,
    columns=["Column 1", "Column 2", "Pearson", "Spearman", "|Difference|"]
)

difference_table = difference_table.sort_values("|Difference|", ascending=False)

display(difference_table.head(3))
```

```python
group_stats = df.groupby("fuel_type")["price"].agg(["mean", "std", "count"])

display(group_stats)

highest_mean = group_stats["mean"].idxmax()
highest_std = group_stats["std"].idxmax()

ratio = group_stats["mean"].max() / group_stats["mean"].min()

print(f"Highest Mean Group: {highest_mean}")
print(f"Highest Std Group: {highest_std}")
print(f"Highest Mean / Lowest Mean Ratio: {ratio:.2f}")
```
